outlier detection


```python
num_cols = df.select_dtypes(include = np.number)
```

```python
num_cols.columns
```

This is the Box Plot method to find the outliers.

```python
Q1 = num_cols.quantile(0.25)
Q3 = num_cols.quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_fence = Q1 - 1.5*IQR
upper_fence = Q3 + 1.5*IQR
```

```python
outlier_mask = ((num_cols < lower_fence) | (num_cols > upper_fence))
```

outlier_mask these are outliers in erery column.
